```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
file = pd.read_csv('Motor_Vehicle_Collisions_-_Crashes.csv', low_memory = False)
```

```python
df = pd.DataFrame(file)
df
```

```python
df = df.sample(n = 10000, replace = True)
df = df.drop(['CONTRIBUTING FACTOR VEHICLE 1','CRASH DATE','BOROUGH','ZIP CODE','LATITUDE','LONGITUDE','LOCATION','ON STREET NAME','CROSS STREET NAME','OFF STREET NAME','CONTRIBUTING FACTOR VEHICLE 2','CONTRIBUTING FACTOR VEHICLE 3','CONTRIBUTING FACTOR VEHICLE 4','CONTRIBUTING FACTOR VEHICLE 5','VEHICLE TYPE CODE 2','VEHICLE TYPE CODE 3','VEHICLE TYPE CODE 4','VEHICLE TYPE CODE 5'], axis = 1)
```

```python
df = df.dropna()
df
```

```python
df = df.drop(['CRASH TIME','NUMBER OF PEDESTRIANS INJURED','NUMBER OF CYCLIST INJURED','NUMBER OF CYCLIST KILLED','COLLISION_ID'], axis = 1)
df
```

```python
def f_feature_selection(data):
    features = data.iloc[:,:-1].values
    observed = data.iloc[:,-1:].values
    return features, observed
```

```python
def f_createTrainTestData(X, y):
    xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.40, random_state=42)
    return xtrain, xtest, ytrain, ytest
```

```python
def f_pairPlot(dataframe):
    sns.pairplot(dataframe, diag_kind='kde', hue='VEHICLE TYPE CODE 1')
```

```python
def f_knn(X_train, X_test, y_train, y_test):
    model_performance = 0.0
    y_test  = np.reshape(y_test, len(y_test), )
    y_train = np.reshape(y_train, len(y_train), )

    model = KNeighborsClassifier(n_neighbors = 3)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    model_performance = accuracy_score(y_test, y_pred)
    return model_performance.round(3)

def f_svm(X_train, X_test, y_train, y_test):
    model_performance = 0.0

    y_test  = np.reshape(y_test, len(y_test), )
    y_train = np.reshape(y_train, len(y_train), )

    # train the model 
    model = SVC()
    model.fit(X_train, y_train)

    # test the model
    y_pred = model.predict(X_test)

    # accuracy results
    model_performance = accuracy_score(y_test, y_pred)
    return model_performance.round(3)

def f_DecisionTree(X_train, X_test, y_train, y_test):
    model_performance = 0.0
    y_test  = np.reshape(y_test, len(y_test), )
    y_train = np.reshape(y_train, len(y_train), )

    model = tree.DecisionTreeClassifier()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    model_performance = accuracy_score(y_test, y_pred)
    return model_performance.round(3)

def f_bayes(X_train, X_test, y_train, y_test):
    model_performance = 0.0

    y_test  = np.reshape(y_test, len(y_test), )
    y_train = np.reshape(y_train, len(y_train), )

  # train the model 
    model = GaussianNB()
    model.fit(X_train, y_train)

  # test the model
    y_pred = model.predict(X_test)
    
  # accuracy results
    model_performance = accuracy_score(y_test, y_pred)
    return model_performance.round(3)
```

```python
def main():
    status = 0

    X_features, y_feature = f_feature_selection(df)
    xtrain, xtest, ytrain, ytest = f_createTrainTestData(X_features, y_feature)

    # train and test each models
    status = f_knn(xtrain, xtest, ytrain, ytest)
    print(' - KNN model performance = ', status)
    
    status =  f_svm(xtrain, xtest, ytrain, ytest)
    print(' - SVM model performance = ', status)
    
    status = f_DecisionTree(xtrain, xtest, ytrain, ytest)
    print(' - DecisionTree model performance = ', status)
    
    status =  f_bayes(xtrain, xtest, ytrain, ytest)
    print(' - Bayes model performance = ', status)

    print('\n-    -    -    -    -  E N D  -    -    -    -    -    -\n')
```

```python
main()
```
